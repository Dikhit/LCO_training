```python
my_product = []
flipkart = rq.get("https://www.flipkart.com/search?q=smart+watches")
if flipkart.status_code == 200:
    soup = bs4.BeautifulSoup(flipkart.text, "html.parser")
    soup.title.get_text()
    title = soup.findAll('div', attrs = {'class' : '_3wU53n'})
    price = soup.findAll('div', attrs={'class' : '_1vC4OE _2rQ-NK'})
    
    for items in title:
        for i in price:
            i = str(i.get_text())
            i = ''.join([x for x in i if x.isdigit()])
            i = int(i)
            
            if i < 5000:
                with open("watch_data.txt", 'a+') as file:
                    file.write("(" + items.get_text() + "," + str(i) + ") \n")
                    file.write('\n')
```
