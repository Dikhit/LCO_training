```python
from tkinter import *
from tkinter import ttk
from PIL import ImageTk, Image


root = Tk()
root.geometry("400x400+40+40")
root.title("Image Viewer App")

global current_pic_number
current_pic_number = 1

def controller(value):
    if value == 'l':
        current_pic_number = current_pic_number + 1
        url = "images/" + str(current_pic_number) + ".jpg"
        print(value, url)
        img = ImageTk.PhotoImage(Image.open(url))
        img_label = Label(root, image = img)
        img_label.pack()

    elif value == 'r':
        current_pic_number = current_pic_number + 1
        url = "images/" + str(current_pic_number) + ".jpg"
        print(value, url)
        img = ImageTk.PhotoImage(Image.open(url))
        img_label = Label(root, image = img)
        img_label.pack()

left_arrow = ttk.Button(root, text = "<<", command=lambda : controller('l')).pack(side = "left")
right_arrow = ttk.Button(root, text = ">>", command=lambda : controller('r')).pack(side = "right")


root.mainloop()
```

```python
root = Tk()
# root.geometry("400x400+40+40")
root.title("Image Viewer App")


def back():
    global img_label
    global button_forward
    global button_back
    

def forward(img_number):
    global img_label
    global button_forward
    global button_back
    
    img_label.grid_forget()
    img_label = Label(image = img_list[img_number-1])
    img_label.grid(row = 0, column = 0, columnspan = 3)
    
    if img_number + 1 <= 5:
        left_button = ttk.Button(root, text = "<<", command = lambda : back(img_number-1)).grid(row = 1, column = 0)
        right_button = ttk.Button(root, text = ">>", command = lambda : forward(img_number+1)).grid(row = 1, column = 2)
    else:
        pass
    
    
    right_button.grid(row = 1, column = 2)
    left_button.grid(row = 1, column = 0)
    
img_1 = ImageTk.PhotoImage(Image.open("images/1.jpg"))
img_2 = ImageTk.PhotoImage(Image.open("images/2.jpg"))
img_3 = ImageTk.PhotoImage(Image.open("images/3.jpg"))
img_4 = ImageTk.PhotoImage(Image.open("images/4.jpg"))
img_5 = ImageTk.PhotoImage(Image.open("images/5.jpg"))


img_list = [img_1,img_2, img_3, img_4, img_5]


img_label = Label(image = img_1,height = 550, width = 600)
img_label.grid(row = 0, column = 0, columnspan = 3)




left_button = ttk.Button(root, text = "<<", command = back).grid(row = 1, column = 0)
right_button = ttk.Button(root, text = ">>", command = lambda : forward(2)).grid(row = 1, column = 2)

root.mainloop()
```
