```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('1.jpg')
```

```python
plt.imshow(img)
```

```python
capture = cv2.VideoCapture(0)
while True:
    _, frame = capture.read()
    cv2.imshow("Frame", frame)
    
    key = cv2.waitKey(1)
    
    if key == 27:
        break
    elif key == ord('s'):
        cv2.imwrite('saved.jpg', frame)
capture.release()
cv2.destroyAllWindows()
```

```python
# gray video capture

capture = cv2.VideoCapture(0)
while True:
    _, frame = capture.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow("Frame", gray)
    
    key = cv2.waitKey(1)
    
    if key == 27:
        break
    elif key == ord('s'):
        cv2.imwrite('saved.jpg', frame)
        
capture.release()
cv2.destroyAllWindows()
```

```python
#video player

capture = cv2.VideoCapture('output.avi')
while capture.isOpened():
    _, frame = capture.read()
    cv2.imshow("Frame", frame)
    
    key = cv2.waitKey(1)
    
    if key == 27:
        break
    elif key == ord('s'):
        cv2.imwrite('saved.jpg', frame)
        
capture.release()
cv2.destroyAllWindows()
```
