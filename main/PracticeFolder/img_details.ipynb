```python
import requests as rq
from IPython.display import Image

def random_users():
    random_user_api = rq.get("https://randomuser.me/api/")
    random_user_api_json = random_user_api.json()
        
    data_container = random_user_api_json['results']
            
    img_file = data_container[0]['picture']['medium']
    print(img_file)
    r = rq.get(img_file)
    print(r)

    with open('random_user.jpg','wb') as f:
        f.write(r.content)

    print(Image(filename="random_user.jpg", width=100))
```

```python
import requests as rq


user_name=input("Enter your Github username: ")
Password = input("Enter your password")


url = rq.get('https://api.github.com/user', auth=(user_name, Password))

data = url.json()

Name = data['login']
Username = data['name']
Location = data['location']
public_repo = data['public_repos']


print('Name: {}'.format(Name))
print('Username: {}'.format(Username))
print('Location: {}'.format(Location))
print('public_repo: {}'.format(public_repo))
```
