```python
import numpy as np 
import cv2  
cap = cv2.VideoCapture(0)  
reshaped_kernel = np.array([[0,0,0],[0,1,0],[0,0,0]]) 
count = 5  
while 1:          
    _, frame = cap.read()     
    output = cv2.filter2D(frame, -1, reshaped_kernel)          
    cv2.imshow("Webcam", output)          
    choice = cv2.waitKey(1)          
    if choice == 27:
        break          
    elif choice == ord('x'):         
        cv2.imwrite("Screenshot.jpg", output)          
    elif choice == ord('b'):
        blur_kernel = np.ones((count), dtype = np.float32)/count         
        reshaped_kernel = blur_kernel.reshape(1,-1)         
        kernel = 'b'              
    elif choice == ord('e'):         
        embossed_kernel = np.array([[-2*count,-1*count,0],[-1*count,count,count],[0,1*count,2*count]])         
        reshaped_kernel = embossed_kernel         
        kernel = 'e'              
    elif choice == ord('s'):         
        sharpen_kernel = np.array([[0,-1*count,0],[-1*count,4*count+1,-1*count],[0,-1*count,0]])         
        reshaped_kernel = sharpen_kernel         
        kernel = 's'              
    elif choice == ord('o'):         
        outline_kernel = np.array([[-1*count,-1*count,-1*count],[-1*count,8*count,-1*count],[-1*count,-1*count,-1*count]])         
        reshaped_kernel = outline_kernel         
        kernel = 'o'              
    elif choice == ord('+'):         
        if kernel == 'b':             
            count += 1             
            blur_kernel = np.ones((count), dtype = np.float32)/count             
            reshaped_kernel = blur_kernel.reshape(1,-1)         
        elif kernel == 'e':             
            count += 1             
            reshaped_kernel = np.array([[-2*count,-1*count,0],[-1*count,count,count],[0,1*count,2*count]])         
        elif kernel == 's':             
            count += 1             
            reshaped_kernel = np.array([[0,-1*count,0],[-1*count,4*count+1,-1*count],[0,-1*count,0]])         
        elif kernel == 'o':             
            count += 1             
            reshaped_kernel = np.array([[-1*count,-1*count,-1*count],[-1*count,8*count,-1*count],[-1*count,-1*count,-1*count]])              
    elif choice == ord('-'):         
        if count > 1:             
            if kernel == 'b':                 
                count -= 1                 
                blur_kernel = np.ones((count), dtype = np.float32)/count                 
                reshaped_kernel = blur_kernel.reshape(1,-1)             
            elif kernel == 'e':                 
                count -= 1                 
                reshaped_kernel = np.array([[-2*count,-1*count,0],[-1*count,count,count],[0,1*count,2*count]])             
            elif kernel == 'o':                 
                count -= 1                 
                reshaped_kernel = np.array([[-1*count,-1*count,-1*count],[-1*count,8*count,-1*count],[-1*count,-1*count,-1*count]])         
            elif count > 5:             
                if kernel == 's':                 
                    count -= 1                 
                    reshaped_kernel = np.array([[0,-1*count,0],[-1*count,4*count+1,-1*count],[0,-1*count,0]])          
cap.release() 
cv2.destroyAllWindows()
```

```python
capture = cv2.VideoCapture(0)
user_kernel = np.array([[0,0,0],[0,1,0],[0,0,0]]) 

effect_degree = 10

while True:
    
#     initialise the webcam
    _, frame = capture.read()
    output_frame = cv2.filter2D(frame, -1, user_kernel) 
    cv2.imshow("Frame", output_frame)
    
#     keystroke capture
    key_stroke = cv2.waitKey(1)
    
    if key_stroke == 27:
        break
    else:
#         blur the frame
        if key_stroke == ord('B'):
            effect_degree += 1             
            blur_kernel = np.ones((effect_degree), dtype = np.float32)/effect_degree
            user_kernel = blur_kernel
            
        elif key_stroke == ord('x'):
            cv2.imwrite("SS.jpg", output_frame)
        
        elif key_stroke == ord('b'):
            if effect_degree > 1:
                effect_degree -= 1
                blur_kernel = np.ones((effect_degree), dtype = np.float32)/effect_degree
                user_kernel = blur_kernel
            else:
                pass
            
#         sharpen kernel
        elif key_stroke == ord('S'):
            effect_degree += 1
            sharpen_kernel = np.array([[0,-1*effect_degree,0],[-1*effect_degree,4*effect_degree+1,-1*effect_degree],[0,-1*effect_degree,0]])  
            user_kernel = sharpen_kernel
        elif key_stroke == ord('s'):
            if effect_degree > 1:
                effect_degree -= 1
                sharpen_kernel = np.array([[0,-1*effect_degree,0],[-1*effect_degree,4*effect_degree+1,-1*effect_degree],[0,-1*effect_degree,0]])  
                user_kernel = sharpen_kernel
            else:
                pass
            
#         outline kernel
        elif key_stroke == ord('O'):
            effect_degree += 1
            outline_kernel = np.array([[-1*effect_degree,-1*effect_degree,-1*effect_degree],[-1*effect_degree,8*effect_degree,-1*effect_degree],[-1*effect_degree,-1*effect_degree,-1*effect_degree]])
            user_kernel = outline_kernel
        elif key_stroke == ord('o'):
            if effect_degree > 1:
                effect_degree -= 1
                outline_kernel = np.array([[-1*effect_degree,-1*effect_degree,-1*effect_degree],[-1*effect_degree,8*effect_degree,-1*effect_degree],[-1*effect_degree,-1*effect_degree,-1*effect_degree]])
                user_kernel = outline_kernel
            else:
                pass
    
capture.release()
cv2.destroyAllWindows()
```
