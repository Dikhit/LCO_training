```python
import cv2 as cv
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import requests
import io

url = "https://image.shutterstock.com/image-photo/beautiful-afro-man-front-brown-260nw-620050766.jpg"
image = Image.open(io.BytesIO(requests.get(url,stream = True).content)).convert("RGB")
```

```python
image
```

```python
image_array = np.array(image)
```

```python
grey_img = cv2.cvtColor(image_array, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(grey_img)
```

```python
cv2.__path__
```

```python
path = "C:\\Users\\Katlic\\AppData\\Local\\Continuum\\anaconda3\\Lib\\site-packages\\cv2\\data\\haarcascade_frontalface_default.XML"
detector = cv2.CascadeClassifier(path)
```

```python
faces = detector.detectMultiScale(grey_img, 1.3,5)
```

```python
faces
```

```python
for face in faces:  
    x,y,w,h = face
    cv.rectangle(image_array, (x,y), (x+w, y+h), (5,220,1000), 5)
    
plt.imshow(image_array)
```

```python

import cv2
import numpy as np
import math

cap = cv2.VideoCapture(0)
    
while True:
    _, frame = cap.read()
    image_array = np.array(frame)
    gray = cv2.cvtColor(image_array, cv2.COLOR_BGR2GRAY)

    path = "C:\\Users\\Katlic\\AppData\\Local\\Continuum\\anaconda3\\Lib\\site-packages\\cv2\\data\\haarcascade_frontalface_default.XML"
    detector = cv2.CascadeClassifier(path)
    faces = detector.detectMultiScale(gray,1.3,5)
    for face in faces:
        x,y,w,h = face
        cv2.rectangle(gray, (x,y), (x+w, y+h), (255,0,0), 5)
    
    cv2.imshow("See your funny face ", gray)
    key_stroke = cv2.waitKey(1)
    if key_stroke == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
